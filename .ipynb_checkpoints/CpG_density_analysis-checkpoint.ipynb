```python
Analysis of the "CpG density" in coding regions of Cosmic cancer genes(those genes known to be involved in cancer according to the Cosmic census) versus all other genes. The CpG density is defined as the # of CpG dinucleotides divided by the total length of the coding region X 100.
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import pickle 
cancerGeneToCpGDensity = pickle.load(open('/Users/khandekara2/Documents/methylationProject/01_data/cancerGeneToCpGDensity.pickle', 'rb'))
noncancerGeneToCpGDensity = pickle.load(open('/Users/khandekara2/Documents/methylationProject/01_data/noncancerGeneToCpGDensity.pickle', 'rb'))
```

```python
print (cancerGeneToCpGDensity)
```

```python
cancer_ratios = [i * 100 for i in cancerGeneToCpGDensity.values()]
noncancer_ratios = [i * 100 for i in noncancerGeneToCpGDensity.values()]
```

```python
plt.hist(cancer_ratios, bins=25)
plt.xlabel('CpG Density')
plt.ylabel('Frequency')
plt.title('Distribution of CpG Densities of Cosmic Cancer Genes')
```

```python
plt.hist(noncancer_ratios, bins=25)
plt.title('Distribution of CpG Densities of Non-cancer Genes')
plt.xlabel('CpG Density')
plt.ylabel('Frequency')
```

```python
print (sum(cancer_ratios) / len(cancer_ratios))
```

```python
print (sum(noncancer_ratios) / len(noncancer_ratios))
```
